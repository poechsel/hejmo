```python
import os
import pandas as pd
import json
import folium
import numpy as np
```

```python
positions_json = pd.read_json("positions.json")
positions_json.head()
```

```python
positions = pd.io.json.json_normalize(positions_json["locations"])
positions.latitudeE7 /= 10000000
positions.longitudeE7 /= 10000000
positions.head()
```

```python
positions['position'] = positions[['latitudeE7', 'longitudeE7']].apply(tuple, axis=1)
```

```python
coordinates_switzerland = [46.8182, 8.2275]
map_world = folium.Map(coordinates_switzerland, tiles='cartodbpositron', zoom_start=8)
map_world
```

```python
positions.sample(100)
```

```python
feature = folium.FeatureGroup("Locations")
sp = positions.sample(100).position
sp.apply(lambda pos : feature.add_child(folium.Marker(pos)))
sp
```

```python
map_world.add_child(feature)
```
